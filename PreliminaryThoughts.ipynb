```python
# Target conditions
target_unstable = [[5,     10,    15],              # WS
                   [0.080, 0.075, 0.09]]            # TI levels
target_neutral  = [[5,     10,    15],              # WS
                   [0.055, 0.055, 0.065]]           # TI levels
target_stable   = [[5,     10,    15],              # WS
                   [0.045, 0.050, 0.06]]           # TI levels
```

```python
from pylab import *
veryus    = loadtxt('Data/Archer_2016_WS_TKE_CW_veryunstable.csv', skiprows=2, delimiter=',')
unstable  = loadtxt('Data/Archer_2016_WS_TKE_CW_unstable.csv', skiprows=2, delimiter=',')
neutral   = loadtxt('Data/Archer_2016_WS_TKE_CW_neutral.csv', skiprows=2, delimiter=',')
stable    = loadtxt('Data/Archer_2016_WS_TKE_CW_stable.csv', skiprows=2, delimiter=',')
verystable= loadtxt('Data/Archer_2016_WS_TKE_CW_verystable.csv', skiprows=2, delimiter=',')

figure(figsize=(12,8), facecolor='white')
plot(veryus[:,0], sqrt(2.0/3.0*veryus[:,1])/veryus[:,0], 'r.', label='very unstable')
plot(unstable[:,0], sqrt(2.0/3.0*unstable[:,1])/unstable[:,0], 'y.', label='unstable')
plot(neutral[:,0], sqrt(2.0/3.0*neutral[:,1])/neutral[:,0], 'g.', label='neutral')
plot(stable[:,0], sqrt(2.0/3.0*stable[:,1])/stable[:,0], 'b.', label='stable')
plot(verystable[:,0], sqrt(2.0/3.0*verystable[:,1])/verystable[:,0], 'k.', label='very stable')

plot(target_unstable[0], target_unstable[1], 'rP', markersize=15, label='target unstable')
plot(target_neutral[0],  target_neutral[1],  'gP',  markersize=15, label='target neutral')
plot(target_stable[0],   target_stable[1], 'bP', markersize=15, label='target stable')
legend()
xlabel('WS [m/s]')
ylabel('TI [-]')
title('TI vs. WS met-mast measurements, z=20m')
tight_layout()
show()
```
